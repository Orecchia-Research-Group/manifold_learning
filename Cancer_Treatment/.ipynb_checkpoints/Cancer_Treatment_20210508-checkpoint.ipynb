## Read in data

```python
# Untreated
untr = pd.read_csv("GSM3972657_D0.dge.txt", sep='\t', header = 0, index_col = 0)

# Erlotinib treated
E01 = pd.read_csv("GSM3972658_D1.dge.txt", sep='\t', header = 0, index_col = 0)
E02 = pd.read_csv("GSM3972659_D2.dge.txt", sep='\t', header = 0, index_col = 0)
E04 = pd.read_csv("GSM3972660_D4.dge.txt", sep='\t', header = 0, index_col = 0)
E09 = pd.read_csv("GSM3972661_D9.dge.txt", sep='\t', header = 0, index_col = 0)
E11 = pd.read_csv("GSM3972662_D11.dge.txt", sep='\t', header = 0, index_col = 0)

# Erlotinib treated with break
E11_D8 = pd.read_csv("GSM3972672_D19DMSO.dge.txt", sep='\t', header = 0, index_col = 0)
E11_D6_E2 = pd.read_csv("GSM3972673_D19_ERL.dge.txt", sep='\t', header = 0, index_col = 0)
```

```python
print("Untr: "+repr(untr.shape))

print("E01: "+repr(E01.shape))
print("E02: "+repr(E02.shape))
print("E04: "+repr(E04.shape))
print("E09: "+repr(E09.shape))
print("E11: "+repr(E11.shape))

print("E11_D8: "+repr(E11_D8.shape))
print("E11_D6_E2: "+repr(E11_D6_E2.shape))
```

```python
def intersection(lst1, lst2):
    return list(set(lst1) & set(lst2))
```

```python
gene_list = intersection(list(untr.index), list(E01.index))
gene_list = intersection(gene_list, list(E02.index))
gene_list = intersection(gene_list, list(E04.index))
gene_list = intersection(gene_list, list(E09.index))
gene_list = intersection(gene_list, list(E11.index))
gene_list = intersection(gene_list, list(E11_D8.index))
gene_list = intersection(gene_list, list(E11_D6_E2.index))

print(len(gene_list))
```

```python
untr = untr.loc[gene_list]

E01 = E01.loc[gene_list]
E02 = E02.loc[gene_list]
E04 = E04.loc[gene_list]
E09 = E09.loc[gene_list]
E11 = E11.loc[gene_list]

E11_D8 = E11_D8.loc[gene_list]
E11_D6_E2 = E11_D6_E2.loc[gene_list]
```

```python
print("Untr: "+repr(untr.shape))

print("E01: "+repr(E01.shape))
print("E02: "+repr(E02.shape))
print("E04: "+repr(E04.shape))
print("E09: "+repr(E09.shape))
print("E11: "+repr(E11.shape))

print("E11_D8: "+repr(E11_D8.shape))
print("E11_D6_E2: "+repr(E11_D6_E2.shape))
```

```python
np_data = np.append(untr.values, E01.values, axis=1)
np_data = np.append(np_data, E02.values, axis=1)
np_data = np.append(np_data, E04.values, axis=1)
np_data = np.append(np_data, E09.values, axis=1)
np_data = np.append(np_data, E11.values, axis=1)
np_data = np.append(np_data, E11_D8.values, axis=1)
np_data = np.append(np_data, E11_D6_E2.values, axis=1)
print(np_data.shape)


cell_id = []
for cell in untr.columns:
    cell_id.append(cell)
    
for cell in E01.columns:
    cell_id.append(cell)
    
for cell in E02.columns:
    cell_id.append(cell)

for cell in E04.columns:
    cell_id.append(cell)
    
for cell in E09.columns:
    cell_id.append(cell)

for cell in E11.columns:
    cell_id.append(cell)

for cell in E11_D8.columns:
    cell_id.append(cell)
    
for cell in E11_D6_E2.columns:
    cell_id.append(cell)
    
print(len(cell_id))

pd_data = pd.DataFrame(np_data, index = gene_list, columns = cell_id)
print(pd_data.shape)
```

## Persistence

```python
# ----------------------- Palettes ------------------------

blue = ["blue", "cornflowerblue", "lightskyblue"]
red = ["maroon", "indianred", "salmon"]
orange = ["orangered","coral", "orange"]
purple = ["indigo", "rebeccapurple", "mediumpurple"]
green = ["darkgreen", "limegreen", "lightgreen"]
yellow = ["darkgoldenrod", "goldenrob", "gold"]
teal = ["darkcyan","lightseagreen","paleturquoise"]

pal_array = np.array([np.array(blue), np.array(red), np.array(orange), np.array(purple), np.array(green), np.array(yellow), np.array(teal)])

custom_palette = pd.DataFrame(pal_array, index = ["blue", "red", "orange", "purple", "green", "yellow", "teal"], columns = ["dark", "medium", "light"])
```

```python
def persistance_diagram(expression_matrix, color, title_plt):
    
    # expression_matrix - expression matrix to be fed into Risper (will be transposed)
    # color - "blue", "red", "orange", "purple", "green", "yellow"
    # title_plt - title for returned persistance diagram
    # returns birth-death pairs for lifetime
    
    bd = ripser(expression_matrix.T, maxdim=2)['dgms']
    
    xy_H_0 = bd[0]
    xy_H_1 = bd[1]

    y_max = sorted(list(xy_H_0[:,1]))[-2]
    print(sorted(list(xy_H_0[:,1]))[-1])
    x_max = sorted(list(xy_H_1[:,1]))[-1]
    either_max = np.amax(np.array([x_max, y_max]))
    
    buffer = either_max*0.1
    
    x = np.linspace(-1*buffer,either_max,500)

    plt.figure(figsize=(5, 5))
            
    sns.scatterplot(xy_H_0[:,0], xy_H_0[:,1], hue = np.ones(xy_H_0.shape[0]), s = 5, linewidth=0, palette = [custom_palette.loc[color]["medium"]], label = "H_0")
    sns.scatterplot(xy_H_1[:,0], xy_H_1[:,1], hue = np.ones(xy_H_1.shape[0]), s = 5, linewidth=0, palette = [custom_palette.loc[color]["dark"]], label = "H_1")
    plt.plot(x, x, ':k', label='Birth = Death')
    plt.plot(x, either_max*np.ones(x.shape), ':k', label='Fully Connected Component')
    plt.ylabel('Death')
    plt.xlabel('Birth')
    plt.ylim((-1*buffer,either_max*1.1))
    plt.xlim((-1*buffer,either_max*1.1))
    plt.title(title_plt)
    plt.legend(bbox_to_anchor=(1.2, 1),borderaxespad=0)
    #plt.savefig('.png')
    plt.show()
    
    return bd, color
```
